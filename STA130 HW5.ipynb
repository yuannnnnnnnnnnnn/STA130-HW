````python
import scipy.stats as stats

def calculate_p_value(num_test, num_right, hypothesized_prob=0.5):
    # Calculate the p-value using the binomial test
    p_value = stats.binom_test(num_right, num_test, hypothesized_prob, alternative='two-sided')
    return p_value

# Example usage
num_test = 124  # total number of tests
num_right = 80   # number of turning right obtained

p_value = calculate_p_value(num_test, num_right)
print(f"P-value: {p_value}")
````

The level of evidence when kissing couples tilt their heads is moderate evidence against the null hypothesis since the p-value is between 0.05 and 0.01.

## **Summary: One-Sided vs. Two-Sided Hypothesis Testing in Vaccine Study**

In our conversation, we explored the differences between one-sided and two-sided hypothesis tests, focusing on how to modify the code for the one-sided test. The key points and adjustments are summarized below:

### **1. Hypothesis Framework**
- **Two-Sided Test**:  
  - **Null Hypothesis (H₀)**: The effect is 0 (no change).  
  - **Alternative Hypothesis (H₁)**: The effect is not equal to 0 (change could be positive or negative).

- **One-Sided Test**:  
  - **Null Hypothesis (H₀)**: The effect is ≤ 0 (no improvement or negative effect).  
  - **Alternative Hypothesis (H₁)**: The effect is > 0 (positive improvement only).  
  - We focused on testing whether the observed health score changes indicated a positive effect of the vaccine.

### **2. Key Code Adjustments for One-Sided Test**
- **Observed Test Statistic Calculation**:  
  ```python
  observed_test_statistic = (patient_data.HealthScoreChange > 0).mean()
  ```
  This remained the same in both tests, calculating the proportion of patients with positive health score changes. However, in the one-sided test, it’s only used to detect improvement (ignoring negative changes).

- **P-Value Calculation Adjustment**:  
  In the one-sided test, the p-value reflects only one direction—whether the observed effect is greater than 0.  

  ```python
  SimTestStats_as_or_more_extreme_than_ObsTestStat = IncreaseProportionSimulations_underH0random >= 0.5
  p_value = (SimTestStats_as_or_more_extreme_than_ObsTestStat).sum() / number_of_simulations
  ```
  This adjustment ignores the negative side and focuses solely on simulated statistics that suggest improvement.

### **3. Interpretation of P-Values**
- **Two-Sided Test**: The p-value considers both extreme positive and negative values, making it more conservative.
- **One-Sided Test**: The p-value only accounts for positive changes, typically resulting in a smaller value (more sensitive to detecting improvement).

### **4. When to Use Each Test**
- **Two-Sided Test**: Use when you are open to detecting changes in either direction (positive or negative).
- **One-Sided Test**: Use when you are only interested in one direction, such as proving a positive impact of the vaccine.

- Differences
    - The sample size of Dr.Bristol is 8 while the sample size of STA130 is 80. 
    - Dr.Bristol's test is an exact test because the sample size is small while it's a z-test for students because the sample size is larger. 
    
Statements of the Null Hypothesis and Alternative hypothesis
- Null hypothesis($H_o$): The students are just guessing, and the probability of correctly identigying the order is 0.5
    - $H_o$ : p=0.5

- Alternative Hypothesis($H_1$): The students can correctly identify the order better than random guessing
    - $H_1$ : p>0.5

Quantitative Analysis

````python
# Input values
n = 80  # Total number of students
X = 49  # Number of correct identifications
p_hat = X / n  # Sample proportion
p_0 = 0.5  # Null hypothesis proportion

# Z-test for proportions formula
z = (p_hat - p_0) / ((p_0 * (1 - p_0) / n) ** 0.5)

# Calculate p-value (one-tailed test, upper tail)
from scipy.stats import norm
p_value = 1 - norm.cdf(z)

# Display the result
if p_value < 0.05:
    print(f"Reject the null hypothesis (z = {z:.2f}, p = {p_value:.4f}).")
else:
    print(f"Fail to reject the null hypothesis (z = {z:.2f}, p = {p_value:.4f}).")
````

By using the students size and number of students who got correct, we can calculate the sample proportion. 

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create the x values for the normal distribution curve
x = np.linspace(0, 1, 1000)
# Create the normal distribution curve
y = norm.pdf(x, loc=p_0, scale=np.sqrt(p_0 * (1 - p_0) / n))

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(['Observed Proportion', 'Expected Proportion'], [p_hat, p_0], color=['blue', 'orange'], alpha=0.7)
plt.ylabel('Proportion')
plt.title('Proportion of Correct Identifications in the Tea Experiment')
plt.ylim(0, 1)
plt.axhline(y=p_0, color='orange', linestyle='--', label='Null Hypothesis (p=0.5)')

# Plot the normal distribution curve
plt.plot(x, y * (max(p_hat, p_0) * n), color='green', label='Normal Distribution Curve')

# Add annotations
plt.text(0, p_hat + 0.02, f'Observed: {p_hat:.2f}', ha='center', color='blue')
plt.text(1, p_0 + 0.02, f'Expected: {p_0:.2f}', ha='center', color='orange')

# Add legend and show plot
plt.legend()
plt.grid(axis='y')
plt.show()
````

In conclusion, we reject the null hypothesis since p-value is lower than the null hypothesis proportion, meaning that the students are not just randomly guessing.
