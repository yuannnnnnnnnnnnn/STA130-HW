```python
import pandas as pd

columns_to_load = ['DEMO_household_income', 'DEMO_height', 'LONELY_ucla_loneliness_scale_companionship', 'CONNECTION_activities_chat_p3m', 'WELLNESS_self_rated_mental_health']
df = pd.read_csv('https://raw.githubusercontent.com/pointOfive/stat130chat130/refs/heads/main/CP/CSCS_data_anon.csv', usecols=columns_to_load)
df
```

```python
df_clean = df.dropna(subset=['LONELY_ucla_loneliness_scale_companionship', 'WELLNESS_self_rated_mental_health'])

df_clean
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Grouped bar plot
sns.countplot(
    data=df_clean,
    x='WELLNESS_self_rated_mental_health',
    hue='LONELY_ucla_loneliness_scale_companionship'
)
plt.title("Relationship Between Loneliness and Mental Health")
plt.show()
```

Null hypothesis -> there's no relationship between loneliness scale and mental helath score

```python
# Filter data for people who feel lonely some of the time

mental_health_order = ['Excellent', 'Very good', 'Good', 'Fair', 'Poor']
some_lonely = df_clean[df_clean['LONELY_ucla_loneliness_scale_companionship'] == 'Often']

# Grouped bar plot for this subset
plt.figure(figsize=(10, 6))  # Adjust the figure size
sns.countplot(
    data=some_lonely,
    x='WELLNESS_self_rated_mental_health',
    hue='LONELY_ucla_loneliness_scale_companionship',
    order=mental_health_order  # Set the order of mental health categories
)
plt.title("Mental Health vs. People Who Often Feel Lonely")
plt.ylabel("Count")
plt.xlabel("Mental Health")
plt.xticks(rotation=30, ha='right')  # Rotate x-axis labels for clarity
plt.legend(title="Loneliness Scale")
plt.tight_layout()  # Ensure everything fits without overlapping
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re

# Example function to convert height to inches
def convert_height_to_inches(height_str):
    # Check if the height string contains inches (e.g., '5' 6"')
    match = re.match(r"(\d+)' (\d+)", height_str)  # Matches height like '5' 6"'
    
    if match:
        feet = int(match.group(1))  # Feet
        inches = int(match.group(2))  # Inches
        return feet * 12 + inches  # Convert feet to inches and add inches
    else:
        # Check if the height contains a number of inches directly (e.g., '60 inches')
        match = re.search(r"(\d+)\s+inches", height_str)
        if match:
            inches = int(match.group(1))
            return inches
        else:
            # If it doesn't match either format, return a NaN (or handle as appropriate)
            return None

# Apply the conversion function to the height column
df_clean['Height_in_inches'] = df_clean['DEMO_height'].apply(convert_height_to_inches)

# Drop rows where height conversion failed (if any)
df_clean = df_clean.dropna(subset=['Height_in_inches'])

# Sort the data by the newly created height column in descending order (tallest to shortest)
df_clean_sorted = df_clean.sort_values(by='Height_in_inches', ascending=False)

# Define the order for mental health categories (flip the order)
mental_health_order = ['Poor', 'Fair', 'Good', 'Very good', 'Excellent']  # Reverse the order

# Create the boxplot with sorted heights
plt.figure(figsize=(10, 6))  # Adjust figure size
sns.boxplot(
    data=df_clean_sorted,
    x='WELLNESS_self_rated_mental_health',  # Mental health categories on x-axis
    y='Height_in_inches',  # Height in inches on y-axis
    order=mental_health_order,  # Ensure reversed order for categories
    palette='viridis'  # Color palette for aesthetic appeal
)
plt.title("Distribution of Height Across Mental Health Categories")
plt.xlabel("Mental Health")
plt.ylabel("Height (inches)")
plt.tight_layout()  # Ensure all elements fit nicely
plt.show()
```
